## FEATURE SELECTION

### WHAT ARE FEATURES (X) ?
#### THESE ARE THE COLOUMNS YOU GIVE TO THE MODEL

```python
x = data[["Open","High","Low","Volume"]]
```

### This is what you want the model to predict

```python
y = data["Close"]
```

## TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split

# this function's job is : DIVIDE X AND Y INTO TRAINING AND TESTING PARTS
```

### TRAINING DATA (80%)  = USED TO TEACH THE MODEL
### TESTING DATA (20%) = USED TO CHECK HOW GOOD THE MODEL PERFORM ON UNSEEN DATA
#### x_train = input used for learning
#### y_train = correct output used for learning
#### x_test = input used for prediction
#### y_test = actual output used for checking prediction 

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

## EVALUATION (CHECK ACCUARACY)

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae = mean_absolute_error(y_test,predictions)
mse = mean_squared_error(y_test,predictions)
r2_score = r2_score(y_test,predictions)
```

#### MAE (Mean Absolute Error) → Average error kitna hai. Jitna chhota, utna achha.
#### MSE (Mean Squared Error) → Error ka squared average. Jitna chhota, utna achha.
#### R² Score → Model kitna sahi predict kar raha hai.
#### 1 ke paas → Excellent model ✅
#### 0.9+ → Bahut achha 👍
#### 0.7–0.9 → Achha 🙂
#### 0.5 se niche → Weak model ❌

```python
print("MAE", mae)
print("MSE", mse)
print("R2_SCORE", r2_score)
```

## FEATURE ENGINEERING 

```python
data['MA_50'] = data['Close'].rolling(50).mean()

data['MA_100'] = data['Close'].rolling(100).mean()

data['Daily_Return'] = data['Close'].pct_change()

data['Volatility'] = data['Daily_Return'].rolling(30).std()
```

```python
#data.dropna(inplace=True)
# because, rolling average creates null values, we have to remove them
```

### RETRAIN THE MODEL WITH NEW FEATURES

```python
X = data[['Open',
          'High',
          'Low',
          'Volume',
          'MA_50',
          'MA_100',
          'Daily_Return',
          'Volatility']]

y = data['Close']
```

## MODEL COMPARE

### 1.. LINEAR REGRESSION MODEL
#### WHICH I HAVE ALLREADY EXPLAIN ABVE AND WE FOUND MAE,,MSE,R2_SCORE

## 2.. DECISION TREE REGRESSOR

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=42)

dt_model.fit(x_train, y_train)

dt_pred = dt_model.predict(x_test)
```

#### CHECK ACCURACY


```python
print("MAE =", mean_absolute_error(y_test, dt_pred))
print("MSE =", mean_squared_error(y_test, dt_pred))
print("R2 Score =", r2_score(y_test, dt_pred))
```

## 3. RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(x_train, y_train)

rf_pred = rf_model.predict(x_test)
```

#### CHECK ACCURACY

```python
print("MAE =", mean_absolute_error(y_test, rf_pred))
print("MSE =", mean_squared_error(y_test, rf_pred))
print("R2 Score =", r2_score(y_test, rf_pred))
```

## 4, hyperparameter tuning of random forest regressor
### this is more professional and more usefull

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

rf_model.fit(x_train, y_train)

rf_pred = rf_model.predict(x_test)
```

```python
print("MAE =", mean_absolute_error(y_test, rf_pred))
print("MSE =", mean_squared_error(y_test, rf_pred))
print("R2 Score =", r2_score(y_test, rf_pred))
```

### CREATE A COMPARISON DATAFRAME

```python
results = {
    "Model": [
        "Linear Regression",
        "Decision Tree",
        "Random Forest",
        "Tuned Random Forest"
    ],

    "MAE": [
        2.57,
        5.35,
        3.84,
        3.79
    ],

    "MSE": [
        13.14,
        65.01,
        30.05,
        28.94
    ],

    "R2 Score": [
        0.99908,
        0.99545,
        0.99790,
        0.99798
    ]
}

comparison_df = pd.DataFrame(results)

print(comparison_df)
```
